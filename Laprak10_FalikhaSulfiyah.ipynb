```python
angka = int(input("Masukkan bilangan: "))
def faktorial(x):
    if x == 1:
        return 1
    else:
        return x * faktorial(x-1)
faktor = faktorial(angka)
print(f"faktorial dari {angka} adalah {faktor}")

```
